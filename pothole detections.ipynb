# Task

```python
data_path = "./dataset/data.yaml"
```

## Train the model

### Subtask:
Initiate the training process using the loaded model and the specified data path.


**Reasoning**:
Initiate the training process using the loaded model and the specified data path for a fixed number of epochs.



```python
results = model.train(data=data_path, epochs=100)
```

**Reasoning**:
The traceback indicates a TypeError because the model is a segmentation model (`yolo11n-seg.pt` implicitly loaded from best.pt) and the dataset is formatted for object detection. To fix this, I need to load a detection model instead of a segmentation model. I will explicitly load a YOLOv8n detection model and then retry the training.



```python
model = YOLO('yolov8n.pt')
results = model.train(data=data_path, epochs=100)
```

```python
results = model.predict(
    source="./dataset/valid/images",
    save=True,
    conf=0.35
)

# Save predictions to Drive
!cp -r runs/detect/predict "$PROJECT/pothole_det_predictions"
```
